### Using the fon api

```python
from fonAPI import FonApi
```

```python
fon = FonApi('5fd66827d5e11f5448a6e9d192030f76348815dab2d69eed')
```

### Tuple of devices to search

```python
brands = ('iphone','samsung')
```

### Serching a list of devices for every brand in de brands list.

```python
phones = []
try: 
    for idx, brand in enumerate(brands):
        phones.append(fon.getdevice(brand))
        print('{}, {}'.format(idx, brand))
except Exception as e:
    print(e.args)
```

### List with all the devices found

```python
phones
```

### Change the list to a dict 

```python
all_devices_information = []
for device_list in phones:
    for device in device_list:
        all_devices_information.append(device)
```

```python
df = pd.DataFrame(all_devices_information)
df.to_csv('all_devices_information.csv', index=False)
```

```python
phone_list = []
for brand_phones in phones:
    try: 
        for phone in brand_phones:
            aux = {}
            aux['name'] = phone['DeviceName']
            aux['brand'] = phone['Brand']
            aux['id'] = ''
            try:
                aux['date'] = int(phone['announced'].split(',')[0])
            except:
                aux['date'] = 0
            phone_list.append(aux)
    except Exception as e:
        print(e.args)
```

```python
phone_list
```

```python
df = pd.DataFrame(phone_list)
df.to_csv('phone_list.csv', index=False)
```

```python
phone_list = list(filter(lambda x : x['date'] > datetime.datetime.now().year - 5, phone_list))
```
